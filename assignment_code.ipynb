```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model_name = 'sentence-transformers/bert-base-nli-mean-tokens'
model = SentenceTransformer(model_name)
```

```python
sentences = []
dataset = pd.read_csv('..\Example_Technical_Skills.csv',skiprows=1,names=['tech_skill'],encoding='utf-8')

dataset = dataset.dropna()

for item in range(len(dataset)):
    sentences.append(dataset.tech_skill[item])

tech_skills_encode = model.encode(sentences)
```

```python
raw_sentences = []
raw_dataset = pd.read_csv('..\Raw_Skills_Dataset.csv',skiprows=1,names=['raw_data'],encoding='utf-8')

raw_dataset = raw_dataset.dropna()

for item in range(len(raw_dataset)):
    raw_sentences.append(raw_dataset.raw_data[item])

raw_data_encode = model.encode(raw_sentences)
```

```python
all_labels = []
ratios = []
for raw in raw_data_encode:

    ratio = np.max(cosine_similarity([raw],tech_skills_encode))
    if ratio >= 0.85:
        all_labels.append('is_tech')
        ratios.append(ratio)
        
        
    elif ratio >= 0.75 and ratio < 0.85:
        all_labels.append('is_soft')
        ratios.append(ratio)
        
        
    else:
        all_labels.append('is_raw')
        ratios.append(ratio)
        
    
    
```

```python
raw_dataset["Label"] = all_labels
raw_dataset["Sim Ratio"] = ratios

raw_dataset.head(15)
```
